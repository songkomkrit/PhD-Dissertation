```python
def get_key(dict, val):
    for k, v in dict.items():
        if v == val:
            return k
    raise Exception(f"No key with value {val}")
```

```python
def import_dict(metadatapath):
    with open(metadatapath) as myfile:
    	indep_contents = myfile.read()
    return json.loads(indep_contents)
```
